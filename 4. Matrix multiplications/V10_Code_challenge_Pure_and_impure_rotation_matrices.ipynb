```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
## pure rotation

# 2D input vector
v = np.array([ [3], [-2] ])

thetas = np.linspace(0, 2*np.pi, 360)
norms = np.zeros(len(thetas))

for i, th in enumerate(thetas):

    A = np.array(
        [ 
            [ 2*math.cos(th), -math.sin(th) ], 
            [   math.sin(th),  math.cos(th) ] 
        ]
    )

    w = A@v
    norms[i] = np.linalg.norm(w)
```

```python
plt.plot(norms, 'b-', linewidth=3)
plt.plot( np.ones(len(thetas))*np.linalg.norm(v), 'r-', linewidth=3)
plt.xlabel('Theta')
plt.ylabel('Norm')
plt.xlim(0, 360)
plt.show()
```
